```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv('house_data.csv')
```

```python
dataset.info()#loading dataset
```

```python
dataset.shape
```

```python
dataset.head(200)
```

```python
#deleting irrelevant column
dataset.drop(['id'], axis = 1, inplace = True)
dataset.head(2)
```

```python
#deleting irrelevant column
dataset.drop(['date'], axis = 1, inplace = True)
dataset.head(2)
```

```python
#deleting irrelevant columns
dataset.drop(['grade'], axis = 1, inplace = True)
dataset.head(2)
```

```python
#checking for null values
dataset.isnull().sum()
```

```python
#analysis
a4_dims = (10,8)
fig, ax = plt.subplots(figsize=a4_dims)
cor = dataset.corr()
sns.heatmap(cor,annot = True,cmap = "YlGnBu")
```

```python
#corellation of price with no of bedrooms
a4_dims = (15,5)
fig, ax = plt.subplots(figsize=a4_dims)
sns.barplot(x = dataset.bedrooms,y = dataset.price)
```

```python
#count of respective no of bedrooms
dataset.groupby('bedrooms').price.agg([len,min,max])
```

```python
#removing outliers
df = dataset[(dataset.bedrooms>0)&(dataset.bedrooms<33)].copy()
```

```python
df.shape
```

```python
#Analysing zipcode feature
df.zipcode.value_counts()
```

```python
#all the zipcodes are of washington.let's do a correlation analysis
```

```python
a4_dims = (1,2)
fig, ax = plt.subplots(figsize=a4_dims)
sns.scatterplot(ax = ax,x = df.price,y = df.zipcode)
```

```python
a4_dims = (15,8)
fig, ax = plt.subplots(figsize=a4_dims)
sns.distplot(a = df.price,bins = 100,color = 'r', ax = ax)
```

```python
#groupby on price
df.price.agg([min,max])
```

```python
len(df[df.price == 0])#there is no instance with price=0
```

```python
#analysis on bathroom wrt price
a4_dims = (15,5)
fig, ax = plt.subplots(figsize=a4_dims)
sns.barplot(x = df.bathrooms,y = df.price)
```

```python
#analysing on all the instances whose price is 0
zero_price = df[(df.price == 0)].copy()
zero_price.shape
```
